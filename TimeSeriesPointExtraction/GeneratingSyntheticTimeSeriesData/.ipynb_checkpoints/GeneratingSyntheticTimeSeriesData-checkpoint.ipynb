```python
# maths
import numpy as np
# personal plotting function
import matplotlib.pyplot as plt
from PlotHelp import plot_time_series
# use os to get path
import os
PROJ_ROOT = os.path.join(os.pardir)
FIG_PATH = PROJ_ROOT + '\\' + 'GeneratingSyntheticTimeSeriesData' + '\\' + 'figures' + '\\'
FIG_PATH
```

## Linear Upward/Downward Trend

```python
# generate data
time = np.arange(100)
values = time*0.2
# call plot function
plot_time_series(time, values, "Upward Trend")
# save 
plt.savefig(FIG_PATH + 'upward_trend'+'.png', transparent=True)
```

```python
# generate data
time = np.arange(100)
values = time[::-1]*0.2
# call plot function
plot_time_series(time, values, "Downward Trend")
# save 
plt.savefig(FIG_PATH + 'downward_trend'+'.png', transparent=True)
```

# Cyclical Trend

```python
# generate data
# create random pattern
time = np.arange(50)
values = np.where(time < 10, time**3, (time-9)**2)
# Repeat pattern 5 times
cyclic = []
for i in range(5):
    for j in range(50):
        cyclic.append(values[j])
time_cyclic = np.arange(250)
# call plot function
plot_time_series(time_cyclic, cyclic, label="Cyclic Trend")
# save 
plt.savefig(FIG_PATH + 'cyclic_trend'+'.png', transparent=True)
```

# Noise

```python
# generate data
noise = np.random.randn(250)*100
cyclic += noise
time_cyclic = np.arange(250)
# call plot function
plot_time_series(time_cyclic, cyclic, label="Cyclic with Noise")
# save 
plt.savefig(FIG_PATH + 'cyclic_noise_trend'+'.png', transparent=True)
```

# Multiple Trends

```python
#generate data
cyclic_upward = cyclic + np.arange(250)*10
time_cyclic = np.arange(250)
# call plot function
plot_time_series(time_cyclic, cyclic_upward, label="Cyclic + Upward Trend + Noise")
# save 
plt.savefig(FIG_PATH + 'upward_cyclic_noise_trend'+'.png', transparent=True)
```

# White Noise

```python
# generate data
time = np.arange(200)
values = np.random.randn(200)*100
# call plot function
plot_time_series(time, values, label="White Noise")
# save 
plt.savefig(FIG_PATH + 'white_noise'+'.png', transparent=True)
```

# Non Stationary Trends

```python
# generate data
big_event = np.zeros(250)
big_event[-50:] = np.arange(50)*-50
non_stationary = cyclic_upward + big_event
time_cyclic = np.arange(250)
# call plot function
plot_time_series(time_cyclic, non_stationary, label="Non-stationary Trends")
# save 
plt.savefig(FIG_PATH + 'non_stationary_trend'+'.png', transparent=True)
```
